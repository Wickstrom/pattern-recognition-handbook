### Unrolling the Swiss roll.

Transform the Swiss roll data into a 2-dimensional representation using Laplacian Eigenmaps


```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_swiss_roll
X, t = make_swiss_roll(n_samples=5000, noise=0.05, random_state=0)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, marker='o')

plt.show()
```
