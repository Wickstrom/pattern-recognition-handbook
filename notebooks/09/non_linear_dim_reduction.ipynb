```python
import numpy as np
import matplotlib.pyplot as plt
from ucimlrepo import fetch_ucirepo
from sklearn.decomposition import PCA

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features.iloc[:, :2]
y = np.zeros(150)
y[50:100] = 1
y[100:150] = 2
y_names = np.unique(iris.data.targets)

pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

plt.figure(1)
for class_i, class_name in enumerate(y_names):
    plt.scatter(X_2d[np.where(y == class_i)[0], 0], X_2d[np.where(y == class_i)[0], 1], label=class_name)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_swiss_roll
X, t = make_swiss_roll(n_samples=5000, noise=0.05, random_state=0)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, marker='o')

plt.show()
```
