```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

from sklearn.decomposition import PCA, KernelPCA

X, y = make_moons(n_samples=200, noise=0.05, random_state=42)

pca = PCA(n_components=2)
x_pca = pca.fit_transform(X)

kpca = KernelPCA(n_components=2, kernel='rbf', gamma=28)
x_kpca = kpca.fit_transform(X)
```

```python
plt.figure(1, figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.title("Original Space")

plt.subplot(1, 3, 2)
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.title("PCA")

plt.subplot(1, 3, 3)
plt.scatter(x_kpca[:, 0], x_kpca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.title("Kernel PCA")

plt.show()
```

### Making the Two-Moons dataset linearly separable

The code below loads a classic synthetic machine learning dataset, the Two Moons dataset, that we have looked at before. Use either kernel PCA or Laplacian Eigenmaps (or both if you wish) to transform the Two-Moons dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from ucimlrepo import fetch_ucirepo
from sklearn.decomposition import PCA

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features.iloc[:, :2]
y = np.zeros(150)
y[50:100] = 1
y[100:150] = 2
y_names = np.unique(iris.data.targets)

pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

plt.figure(1)
for class_i, class_name in enumerate(y_names):
    plt.scatter(X_2d[np.where(y == class_i)[0], 0], X_2d[np.where(y == class_i)[0], 1], label=class_name)
plt.legend()
plt.show()
```
